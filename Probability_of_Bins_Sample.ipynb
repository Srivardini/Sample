# **Propability of bins**
T_prob: Each bin of Teff constitutes a probablity of the distribution which can be assigned to all the stars within the respective bin. 


```python
data      = df['Teff']
bin_min   = data.min()
bin_max   = data.max()
bin_width = 200

bins = np.arange(bin_min, bin_max+2*bin_width, bin_width)
bins = df['Teff'].value_counts(bins = bins, normalize=True)

# P(T)
T_prob = []
for i,row in df.iterrows():
  flag = 1
  for bounds, val in bins.iteritems():
    if float(row['Teff'])>bounds.left and float(row['Teff'])<=bounds.right:
      T_prob.append(val)
      flag = 0
  if flag:
    print(row)
################################################################################

#P(Met)
data      = df['[Fe/H]']
bin_min   = data.min()
bin_max   = data.max()
bin_width = 0.1

bins = np.arange(bin_min, bin_max+2*bin_width, bin_width)
bins = df['[Fe/H]'].value_counts(bins = bins, normalize=True)

M_prob = []
for i,row in df.iterrows():
  for bounds, val in bins.iteritems():
    if float(row['[Fe/H]'])>bounds.left and float(row['[Fe/H]'])<=bounds.right:
      M_prob.append(val)
################################################################################

#P(Logg)
data      = df['logg']
bin_min   = data.min()
bin_max   = data.max()
bin_width = 0.1

bins = np.arange(bin_min, bin_max+2*bin_width, bin_width)
bins = df['logg'].value_counts(bins = bins, normalize=True)

g_prob = []
for i,row in df.iterrows():
  for bounds, val in bins.iteritems():
    if float(row['logg'])>bounds.left and float(row['logg'])<=bounds.right:
      g_prob.append(val)
```

We add T_prob, g_prob and M_prob in the dataframe.

```python
df['T_prob'] = T_prob
df['logg_prob'] = g_prob
df['Fe_H_prob'] = M_prob

# P(T)*P(Logg)*P(Met)

df['combined'] = df['T_prob']*df['logg_prob']*df['Fe_H_prob']
```

# **Sampling from a 3D distribution:**

Since, we consider three parameters, Teff, g and metallicity for sampling the stars.
We can find the cummulative probabiltiy of T_prob, g_prob and M_prob for each stars and use that as a weight factor when we sample 50 stars from 1111 stars.

```python
df_sample = df.sample(50,weights = df['combined'])
```
